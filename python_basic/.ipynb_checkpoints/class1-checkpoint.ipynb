````python
if condition: # scope
    do_something() # statement
    do_something() # statement
````

````python
if 2 < 1:
    print("hello world")
    print("hello world2")
print("end if")
````

````python
temp = 25
if temp < 25 and temp > 15:
    print("")
else:
    print("")
````
