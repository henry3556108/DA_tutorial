## 什麼是 function

想像一下

function 就是一個會幫你做事情的一個工人

他怎麼做事情的你不一定要知道，只需要知道他工作時需要什麼東西，以及它可以幫助你做什麼事情就好了


舉例來說

```
如果有一個工人，你能夠給他一串紙條

該工人會唸出該紙條的內容

在 python 之中他就是個 print function
```

- 上述的例子之中，這個 print 這個工人(function)，需要的東西就是你要給他的資料，並且他會將你的資料列印在螢幕上(他沒有告訴你任何東西，只是將資料列印在螢幕上)


例子2

```
如果有一個工人，你給他一筆資料

他會告訴你這個資料的長度是多少

在 python 之中他就是 len function
```

- 上述的例子之中，這個 len 這個工人(function)，需要的東西就是你要給他的資料，並且他會將資料的長度告訴你(return)給妳(沒有列印出來，但是有告訴你結果)

## 參數引數


### 參數

- 參數是 function 定義中的一部分。它用於指定 function 需要輸入的資料。

### 引數

- 引數是呼叫 function 時傳遞給 function 的資料。

```python

def introduce(name, age): # parmeter 參數
    print(f"Hello, my name is {name}, my age is {age}")

introduce("Henry", 26) # arguments 引數
```

````python
def introduce(name, age): # parmeter 參數
    print(f"Hello, my name is {name}, my age is {age}")

result = introduce("Henry", 26) # arguments 引數
print(result)
````

## 練習
- 判定一個數值是否為質數
- 直接寫一個函數，接受一個數字，判斷是否為質數

### 範例
```python 
is_prime(2) # 預期回傳值：True
is_prime(4) # 預期回傳值：False
```

````python
def is_prime(target: int) -> bool:
    '''
    judge weather the target is prime number
    '''
    for num in range(2, target):
        if target % num == 0:
            return False
    return True

result = is_prime(3)
print(f"result:{result}")
````

### function

method 是物件的一個屬性。它是一個 function，但它只能由物件本身呼叫。
