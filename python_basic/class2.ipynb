# 常用的 Python 字串 method

## 大小寫轉換

* `str.upper()`：將字串轉換為全大寫
* `str.lower()`：將字串轉換為全小寫
* `str.title()`：將字串轉換為標題格式，每個單字的首字母變為大寫
* `str.capitalize()`：將字串首字母轉換為大寫

## 字串判斷
- **會回傳 bool 給你，這個部分很常與 if else 的語法一起串接使用**

* `str.isalpha()`：判斷字串是否只包含字母
* `str.isdigit()`：判斷字串是否只包含數字
* `str.isalnum()`：判斷字串是否只包含字母和數字
* `str.islower()`：判斷字串是否只包含小寫字母
* `str.isupper()`：判斷字串是否只包含大寫字母
* `str.istitle()`：判斷字串是否為標題格式
* `str.isspace()`：判斷字串是否只包含空白字元

## 字串搜尋

* `str.find(sub, start=0, end=len(string))`: 搜尋子字串第一次出現的位置
* `str.rfind(sub, start=0, end=len(string))`: 搜尋子字串最後一次出現的位置
* `str.count(sub, start=0, end=len(string))`: 計算子字串出現的次數
* `str.startswith(prefix, start=0, end=len(string))`: 判斷字串是否以指定字串開頭
* `str.endswith(suffix, start=0, end=len(string))`: 判斷字串是否以指定字串結尾

## 字串分割與合併

* `str.split(sep=None, maxsplit=-1)`: 以指定分隔符將字串分割為列表
* `str.join(iterable)`: 以指定分隔符將序列合併為字串

## 字串替換

* `str.replace(old, new, count=-1)`: 將字串中的舊字串替換為新字串

## 字串擷取

* `str[index]`：擷取指定索引位置的字元
* `str[start:end]`：擷取指定範圍的字串

## 其他常用 method

* `str.strip(chars=None)`: 去除字串兩端的空白字元或指定字元
* `str.lstrip(chars=None)`: 去除字串左端的空白字元或指定字元
* `str.rstrip(chars=None)`: 去除字串右端的空白字元或指定字元
* `str.format(**mapping)`: 使用格式說明符格式化字串

## ASCII
- ASCII 是美國資訊交換標準程式碼（American Standard Code for Information Interchange）的縮寫，是一種電腦編碼系統，用於表示現代英語和其他西歐語言。它使用 7 個位元組來表示一個字元，因此可以表示 128 個不同的字元，包括大寫和小寫字母、數字、標點符號和控制字元。

### 範例
```

```

## list
- 有序性（Ordering）： 列表中的元素按照添加的順序排列，您可以根據索引位置存取元素。

- 可變性（Mutability）： 列表的元素可以修改，您可以更改、添加和刪除列表中的元素。

- 多樣性（Heterogeneity）： 列表可以包含不同類型的元素，例如數字、字串、布林值等。

- 容器性（Container）： 列表可以包含其他的列表，創建多維列表用於存儲表格狀數據。

- 動態性（Dynamic）： 您可以根據需要隨時增加或減少列表的大小。

- 索引和切片（Indexing and Slicing）： 您可以通過索引位置存取列表中的元素，也可以使用切片操作選取部分元素。

`````python
# 使用方括號創建 list
list_1 = [1, 2, 3]

# 使用 list() 函數創建 list
list_2 = list([4, 5, 6])
`````

### 取值

`````python
# 取 list 中指定索引位置的元素
print(list_1[0])  # 1

# 取 list 中指定範圍的元素
print(list_1[1:3])  # [2, 3]
`````

## 修改

`````python
# 修改 list 中指定索引位置的元素
list_1[0] = 10

print(list_1)

# 修改 list 中指定範圍的元素
list_1[1:3] = [20, 30]

print(list_1)
`````

## 運算
