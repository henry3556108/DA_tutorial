## Python 的第一個程式碼

### “Hello, world！” 是 Python 的第一個程式碼，也是所有 Python 程式設計師的必經之路。



```print("Hello, world!")``` 這個程式碼會在螢幕上輸出以下內容：

```Hello, world!```

````python
print("hello world!") # 這個是我的第一份 code
````

## 為什麼要學習這個程式碼？

- 學習這個程式碼可以幫助我們了解 Python 的基本語法，例如：
    - 如何使用 print() 函數輸出內容
    - 如何使用引號括起字串
    - 如何理解這個程式碼？

- 我們可以把這個程式碼分成兩部分：
1. print() 函數：print() 函數可以用來輸出內容。
2. 字串：字串是一串文字，可以用引號括起。

````python
print("遊戲名稱", "Python 遊戲") # 透過逗號分隔資料
````

````python
print("遊戲名稱", "Python 遊戲", sep=":") # 透過逗號分隔資料
````

### print 這個 function 中，逗號可以用來分隔多個內容。

````python
print("遊戲名稱：", "Python 遊戲")
print("遊戲版本：", "1.0")
print("遊戲開發者：", "小明")
````

### 預設的情況下，每一個 print function 都會換行

````python
print("遊戲名稱：", "Python 遊戲", end="-")
print("遊戲版本：", "1.0", end="-")
print("遊戲開發者：", "小明", end="-")
````

## Operand

- Operand 是運算元，用來表示運算的物件。例如：

1. 數字
2. 字串
3. 變數

### 舉例來說

在運算式``` 1 + 2 ```中：
```
+ 是運算子
1 和 2 是運算元
```

## 補充 複合賦值運算子
簡單來說就是在一個算數運算子後面加上一個 "="，就會形成那個運算子的賦值運算子

```python
gold = 100
gold += 10 # gold => 110
```

## Operator 練習

## 額外資訊 

* 在 Python 裡，可以使用 `int()` 函數將浮點數或者特定的字串轉換為整數。
* 在 Python 裡，可以使用 `float()` 函數將整數轉換為浮點數。
* 在 Python 裡，可以使用 `str()` 函數將資料轉換為字串。

- 以上這樣的形式，從 A 資料型態轉換成 B 資料型態的我們都稱之為 **casting 強制轉換型態**
- 當然 casting 不只有上面三個可以轉換，我們之後會看到更多例子


````python
result = 10
print(result, type(result))

result = str(result)
print(result, type(result))
````

# input 函數

- input 函數，可以用來接收使用者輸入文字或數字。

## input 函數的用法

- 提示文字(可選)
- 函數會以 **字串** 返回使用者輸入的內容。

## input 函數的重要性

input 函數是程式設計中的重要概念，理解它對於編寫互動式程式非常重要。


````python
# 沒有提示詞的 input
name = input()
print(name, type(name))

# 有提示詞的 input
name = input("輸入你的名字：")
print(name, type(name))
````

````python
name = input("輸入你的名字：")
age = input("輸入你的年紀：")
print("我的名字是：", name, "年紀是：", age)
````

## 重要範例
- 先不要執行以下程式，我們一起來看一看，下面兩個程式的執行會有什麼不一樣的差別呢？

````python
# 詢問使用者輸入第一個數字
num1 = input("請輸入第一個數字：")

# 詢問使用者輸入第二個數字
num2 = input("請輸入第二個數字：")

# 計算兩個數字的和
sum = num1 + num2

# 輸出兩個數字的和
print("兩個數字的和是：", sum)
````

````python
# 詢問使用者輸入第一個數字
num1 = int(input("請輸入第一個數字："))

# 詢問使用者輸入第二個數字
num2 = int(input("請輸入第二個數字："))

# 計算兩個數字的和
sum = num1 + num2

# 輸出兩個數字的和
print("兩個數字的和是：", sum)
````

## if else
- 執行區域 執行區塊

````python
if condition: # scope
    do_something() # statement
    do_something() # statement
````

````python
if 2 < 1:
    print("hello world")
    print("hello world2")
print("end if")
````

````python
temp = 25
if temp < 25 and temp > 15:
    print("天氣涼爽")
else:
    print("天氣不涼爽")
````

````python
temp = 10
if temp < 25 and temp > 15:
    print("天氣涼爽")
elif temp <= 15:
    print("天氣寒冷")
elif temp > 25:
    print("天氣酷熱")
else:
    print("")
````
