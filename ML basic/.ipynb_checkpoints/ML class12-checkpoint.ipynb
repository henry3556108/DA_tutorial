### 樹基模型（Tree-based Models）入門課程

#### 1. 決策樹介紹
決策樹是一種常見的機器學習模型，用於分類和回歸問題。決策樹通過一系列的問題和條件分支來進行決策，每個節點表示一個特徵，每個分支代表該特徵的可能值，每個葉子節點代表決策結果或預測值。

**特點：**
- 易於理解和解釋

#### 2. sklearn 的分類決策樹使用範例
使用 `sklearn` 庫中的 `DecisionTreeClassifier` 來構建和訓練分類決策樹。

```python
# 導入必要的庫
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 加載數據集
iris = load_iris()
X = iris.data
y = iris.target

# 分割數據集為訓練集和測試集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# 創建分類決策樹模型
clf = DecisionTreeClassifier()

# 訓練模型
clf.fit(X_train, y_train)

# 預測
y_pred = clf.predict(X_test)

# 評估模型
accuracy = accuracy_score(y_test, y_pred)
print(f'模型準確率: {accuracy:.2f}')
```

#### 3. 回歸樹介紹
回歸樹是決策樹的一種變體，用於回歸問題，即預測連續值。回歸樹的每個葉子節點代表一個數值預測。

**特點：**
- 可以處理非線性數據
- 易於理解和解釋

#### 4. 回歸樹與分類決策樹的差異
- **目標變量**：分類決策樹的目標變量是類別標籤，而回歸樹的目標變量是連續值。
- **損失函數**：分類決策樹使用的損失函數通常是基尼不純度或信息增益，而回歸樹使用的是均方誤差（MSE）。
- **預測輸出**：分類決策樹的預測輸出是類別標籤，回歸樹的預測輸出是數值。

#### 5. sklearn 回歸樹的範例
使用 `sklearn` 庫中的 `DecisionTreeRegressor` 來構建和訓練回歸樹。

```python
# 導入必要的庫
from sklearn.datasets import fetch_california_housing
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# 加載數據集
california = fetch_california_housing()
X = california.data
y = california.target

# 分割數據集為訓練集和測試集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 創建回歸樹模型
reg = DecisionTreeRegressor()

# 訓練模型
reg.fit(X_train, y_train)

# 預測
y_pred = reg.predict(X_test)

# 評估模型
mse = mean_squared_error(y_test, y_pred)
print(f'均方誤差: {mse:.2f}')
print(y_test[:5], y_pred[:5])
```

#### 6. 使用決策樹類型模型的技巧與注意事項
- **過擬合**：決策樹容易過擬合，特別是當樹深度過大時，可以通過修剪、設置最大深度、最小樣本分裂數等參數來控制過擬合。
- **特徵重要性**：決策樹可以計算特徵的重要性，有助於進行特徵選擇和理解模型。
- **數據前處理**：決策樹對數據的標準化要求不高，但處理缺失值和類別特徵可能需要一些預處理。
- **模型解釋性**：決策樹的可視化可以幫助理解模型的決策過程，可以使用 `sklearn.tree.plot_tree` 來可視化樹結構。



```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# 可視化決策樹
plt.figure(figsize=(20,10))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```
