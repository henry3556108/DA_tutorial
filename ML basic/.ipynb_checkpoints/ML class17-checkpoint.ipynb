## 主成分分析（PCA, Principal Component Analysis）

- [非監督式學習 李宏毅](https://www.youtube.com/watch?v=iwh5o_M4BNU)
## 特點：
- 捕捉數據的最大變異性：PCA 將數據映射到一組新的軸（稱為主成分），這些軸依據數據的變異性進行排序。第一主成分捕捉了數據中變異性最大的部分，第二主成分捕捉了次大的變異性，以此類推。
- 無監督學習方法：PCA 不考慮數據的類別標籤，因此它主要用於無監督學習任務。
- 降低維度但保持信息：PCA 通常用於將高維數據降維，這樣可以減少計算負擔，同時儘量保留原數據中的信息。

## 使用情境：
- 數據可視化：當我們有高維數據（例如100維），可以使用 PCA 將數據降到2維或3維，然後進行可視化，幫助理解數據的結構。
- 噪音去除：在某些情況下，高維數據中可能包含噪音，PCA 可以用來去除那些與主要變異無關的噪音。
- 特徵降維：當數據維度過高，模型可能會過擬合，此時可以用 PCA 降維來簡化模型。

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# 創建一個隨機的三維數據集
np.random.seed(42)
data = np.random.rand(300, 3)

# 應用 PCA 將數據降到二維
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data)
# print(singular_values_)
print('explained_variance_ratio:', pca.explained_variance_ratio_) # 成分比例
print('components:', pca.components_) # 成分的投影軸向量，又稱特徵向量（eigenvectors）

# 可視化降維後的數據
plt.scatter(data_pca[:, 0], data_pca[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA on 3D Data')
plt.show()
```

### 使用技巧：
- 標準化數據：在應用 PCA 之前，應該對數據進行標準化（均值為0，方差為1），特別是當不同特徵的量綱不一致時。否則，PCA 可能會傾向於方差較大的特徵。
- 選擇主成分數量：通常根據累積方差解釋率來選擇保留的主成分數量。例如，選擇使累積方差解釋率達到90%或95%的主成分數量。
- 可視化主成分：在降維後，可以使用散點圖或其他可視化工具來檢查主成分是否能夠有效區分數據結構。
### 注意事項：
- PCA 是線性方法：PCA 假設數據的變異性是線性的，對於高度非線性的數據可能效果不佳。在這種情況下，可能需要使用核PCA或其他非線性降維方法。
- 解釋性問題：主成分是線性組合，通常難以解釋。可能需要通過查看負載矩陣來理解每個主成分的重要性。
- 信息損失：PCA 會丟失一些信息，特別是在保留少量主成分時，因此需要權衡信息損失和降維效果。


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.decomposition import KernelPCA
from sklearn.preprocessing import PolynomialFeatures

# 生成玩具數據集
X, y = make_moons(n_samples=100, noise=0.1, random_state=42)

# 生成多項式特徵，將數據擴展到更高維度（例如六維）
poly = PolynomialFeatures(degree=3, include_bias=False)
X_high_dim = poly.fit_transform(X)

# 使用 KernelPCA 將高維數據降維到二維
kpca = KernelPCA(n_components=2, kernel='rbf', gamma=5)  # RBF 核
X_kpca = kpca.fit_transform(X_high_dim)

# 可視化降維後的數據
plt.figure(figsize=(8, 6))
plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c=y, cmap='plasma', edgecolor='k')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Kernel PCA on High-dimensional make_moons Dataset')
plt.colorbar(label='Class')
plt.show()
```

## 因子分析
因子分析的主要目的是找出一組潛在的隱藏變量（因子），這些因子能夠解釋觀測數據中的大部分變異性。與主成分分析（PCA）類似，因子分析試圖將高維數據降維到一個低維空間，但因子分析的模型假設與 PCA 不同。

在因子分析中，觀測變量被建模為潛在因子的線性組合，加上一些不可觀測的誤差。這意味著因子分析不僅試圖找到描述數據結構的隱含因子，還考慮到每個觀測變量中的噪聲。

### FactorAnalysis 的使用場景
- 特徵提取：在高維數據集中，因子分析可以用於提取少數潛在因子，這些因子概括了數據中的大部分變異性。
- 降維：因子分析能夠將數據從高維降維到一個低維空間，同時保留數據中的主要變異性。
- 數據探索：因子分析常用於探索性數據分析，幫助研究者理解觀測變量之間的潛在結構。




### 特點與注意事項
- 模型假設：因子分析假設觀測變量可以由潛在因子的線性組合加上一些噪聲來表示。這與 PCA 不同，後者假設數據的主要變異性來自於觀測變量的線性組合。
- 適合的數據：因子分析通常用於具有**相關性強**的高維數據集中，反之如果特徵之間的相關性不強，就沒有必要使用因子分析，特別是在心理學、社會學等領域的問卷調查中，用於找出潛在的心理因素或性格特徵。
- 多數據類型：因子分析一般適用於連續數據。如果數據是類別型的，可能需要考慮其他技術，如主成分分析或對應分析。

```python
from sklearn.datasets import load_digits
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
X, y = load_digits(return_X_y=True)
rfc = RandomForestClassifier()

scroe = cross_val_score(rfc, X, y, cv=10)
scroe.mean()
```

## Autoencoder 
- Autoencoder 是一種人工神經網絡，主要用於無監督學習中的數據壓縮和特徵學習。其目的是學習一個簡化的數據表示（即編碼），使得通過這個簡化的表示能夠重建出原始數據。

### Autoencoder 組成

- 編碼器（Encoder）：將輸入數據映射到一個低維的隱藏層表示（通常是高維數據的壓縮版本）。
- 解碼器（Decoder）：從這個低維的隱藏層表示重建出原始數據，目標是使重建的數據與原始數據盡可能相似。
#### 特點
- 無監督學習：Autoencoder 是一種無監督學習算法，它不需要標籤數據即可進行訓練。模型學習如何在保持數據重要特徵的同時進行壓縮和重建。
- 數據降維：Autoencoder 通常用於將高維數據壓縮到低維空間，類似於主成分分析（PCA），但 Autoencoder 能夠捕捉非線性關係。
- 特徵學習：通過自動編碼器的訓練，隱藏層可以學習到數據中有意義的特徵，這些特徵可以用於其他任務（如分類）。
- 可擴展性：Autoencoder 結構靈活，可以通過添加更多層來增加模型的表達能力，適應複雜數據。
### 使用場景
- 數據降維：將高維數據（如圖像、文本特徵）降維到較低維度，便於可視化或後續的機器學習任務。
- 數據去噪：訓練一個去噪自動編碼器（Denoising Autoencoder），該模型可以從受噪聲影響的數據中重建干淨的數據。
- 異常檢測：通過訓練 Autoencoder 來學習正常數據的表示，如果某個輸入數據無法被良好地重建，則可能是一個異常數據。

[延伸閱讀](https://jason-chen-1992.weebly.com/home/-autoencoder)

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense

# 加载 MNIST 数据集
(x_train, _), (x_test, _) = mnist.load_data()

# 将数据进行预处理，将像素值缩放到 [0, 1] 并展平为 1D 向量
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
x_train = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
x_test = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))

# 自动编码器的维度设置
input_dim = x_train.shape[1]  # 784, 将 28x28 的图像展平成 784 维的向量
encoding_dim = 32  # 将数据压缩到 32 维

# 定义自动编码器的结构
input_img = Input(shape=(input_dim,))

# 编码器部分
encoded = Dense(128, activation='relu')(input_img)
encoded = Dense(64, activation='relu')(encoded)
encoded = Dense(32, activation='relu')(encoded)  # 更高维度的中间层
encoded_output = Dense(encoding_dim, activation='relu')(encoded)

# 解码器部分
decoded = Dense(32, activation='relu')(encoded_output)  # 与编码器对称的层
decoded = Dense(64, activation='relu')(decoded)
decoded = Dense(128, activation='relu')(decoded)
decoded_output = Dense(input_dim, activation='sigmoid')(decoded)

# 定义完整的自动编码器模型
autoencoder = Model(input_img, decoded_output)

# 定义编码器模型
encoder = Model(input_img, encoded_output)

# 编译模型
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

# 训练自动编码器
autoencoder.fit(x_train, x_train,
                epochs=50,
                batch_size=256,
                shuffle=True,
                validation_data=(x_test, x_test))

# 使用训练好的编码器来进行降维
x_test_encoded = encoder.predict(x_test)
# autoencoder.predict(x_test[0])
```

## 線性判別分析（LDA, Linear Discriminant Analysis）
### 特點：
- 考慮類別標籤：與 PCA 不同，LDA 考慮了數據的類別標籤，並試圖找到一個投影方向，使得不同類別之間的區別最大化。
- 分類專用的降維技術：LDA 專門用於分類問題，旨在將不同類別的樣本分隔開來，使得類間距離最大化，類內距離最小化。
- 最多生成 𝑘−1 個投影方向：如果有 k 個類別，LDA 最多能生成 k−1 個投影方向。
### 使用情境：
- 分類問題：當我們有標籤數據並且需要進行分類時，LDA 可以用來降維並提高分類器的性能。例如，在面部識別任務中，LDA 可以幫助找出能夠區分不同個體的關鍵特徵。
- 特徵選擇：在進行分類任務時，可以使用 LDA 來選擇最具區分力的特徵，從而提高模型的準確性。

```python
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import matplotlib.pyplot as plt

# Load the iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Apply PCA to reduce dimensions to 2
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Apply LDA to reduce dimensions to 2
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X, y)

# Plot the PCA results
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
for target, color in zip([0, 1, 2], ['r', 'g', 'b']):
    plt.scatter(X_pca[y == target, 0], X_pca[y == target, 1], alpha=0.8, color=color, label=iris.target_names[target])
plt.title('PCA of IRIS dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()

# Plot the LDA results
plt.subplot(1, 2, 2)
for target, color in zip([0, 1, 2], ['r', 'g', 'b']):
    plt.scatter(X_lda[y == target, 0], X_lda[y == target, 1], alpha=0.8, color=color, label=iris.target_names[target])
plt.title('LDA of IRIS dataset')
plt.xlabel('LD 1')
plt.ylabel('LD 2')
plt.legend()

plt.show()
print(f"pca: {pca.explained_variance_ratio_.cumsum()}")
print(f"lda: {lda.explained_variance_ratio_.cumsum()}")
```

### 使用技巧：
- 類別均衡：如果數據類別不平衡，可以考慮對數據進行過採樣或下採樣，以提高模型的表現。
- 特徵選擇：在使用 LDA 之前，最好先進行特徵選擇，因為 LDA 對高維數據比較敏感。選擇與類別相關性強的特徵有助於提高分類效果。
- 正則化：如果特徵數量接近或大於樣本數，可能需要對 LDA 進行正則化（如使用 Shrinkage LDA），以防止過擬合。
### 注意事項：
- 類別數量限制：LDA 能夠生成的投影方向最多為 k−1 個（k 是類別數量）。因此，當類別數量較少時，LDA 的降維能力有限。
- 異常點敏感：LDA 對異常點敏感，因為異常點會影響計算的類間和類內散布矩陣。可以考慮在應用 LDA 前進行異常點檢測和處理。

## 獨立成分分析（ICA, Independent Component Analysis）
### 特點：
- 信號分離：ICA 的主要目的是將觀測到的信號分解為一組獨立的源信號。這些源信號之間在統計上是互相獨立的。
- 非高斯性：ICA 假設源信號具有非高斯性，這與 PCA 基於的高斯分佈假設不同。
- 重建原始信號：ICA 可以從多個混合信號中分離出原始的獨立信號源。
### 使用情境：
- 盲源分離（Blind Source Separation）：當我們觀測到多個信號的混合，但不知道每個源信號是什麼時，可以用 ICA 來分離它們。經典例子是雞尾酒會問題，在多個麥克風錄音中分離出每個人的聲音。
- 圖像處理：在圖像處理中，ICA 可用於分離不同的獨立成分，例如分離出影像中的不同層次或結構。
- 神經科學：ICA 被用於分析腦電波（EEG）數據，幫助分離腦信號中的不同成分。

[他講的太好了 我輸了](https://www.youtube.com/watch?v=3-mhd5C-cpU)

```python
import numpy as np
from sklearn.decomposition import FastICA
import matplotlib.pyplot as plt

# 創建兩個獨立信號
# np.random.seed(42)
s1 = np.sin(2 * np.pi * np.linspace(0, 1, 200))
s2 = np.sign(np.sin(3 * np.pi * np.linspace(0, 1, 200)))

# 混合信號
S = np.c_[s1, s2]
A = np.array([[1, 0.5], [0.5, 1]])  # 混合矩陣
X = np.dot(S, A.T)  # 觀測到的信號
# 使用 ICA 分離混合信號
ica = FastICA(n_components=2)
S_ica = ica.fit_transform(X)

# 可視化原始信號、混合信號和分離信號
plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.title('Original Signals')
plt.plot(S)

plt.subplot(3, 1, 2)
plt.title('Mixed Signals')
plt.plot(X)

plt.subplot(3, 1, 3)
plt.title('Separated Signals using ICA')
plt.plot(S_ica)

plt.tight_layout()
plt.show()
```

### 使用技巧：
- 數據中心化和白化：在應用 ICA 之前，應對數據進行中心化（均值為0）和白化處理（使各分量無相關且方差為1），這有助於提高 ICA 的效果和收斂速度。
- 多次運行：由於 ICA 的隨機性，最好多次運行 ICA 並比較結果，確保獲得穩定和一致的獨立成分。
### 注意事項：
- 獨立性假設：ICA 假設信號源是統計上獨立的，這在實際應用中可能不總是成立。如果源信號之間有較強的依賴性，ICA 的效果會下降。
- 混合信號數量：ICA 的性能取決於混合信號的數量。如果混合信號的數量少於獨立源信號的數量，ICA 將無法完全分離出所有源信號。
- 算法敏感性：ICA 對初始條件和算法參數非常敏感，因此不同的算法和參數選擇可能會導致不同的分解結果。


## [t-SNE](https://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html)
t-SNE（t-Distributed Stochastic Neighbor Embedding）是一種常用於資料降維和視覺化的非線性技術。它特別適合將高維度的資料轉換成低維度（通常是2D或3D）來進行資料的視覺化，讓人們更容易理解資料的內部結構和相似性。

t-SNE 主要目的是保留資料的局部結構，也就是在高維度中彼此接近的資料點在低維度中也保持接近。這和一些傳統降維方法（例如PCA）不同，因為PCA關注的是整體的線性結構，而t-SNE更專注於局部資料點的相似性

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

# 1. 載入數據集（digits 是 8x8 像素的手寫數字影像）
digits = datasets.load_digits()
X = digits.data
y = digits.target

# 2. 標準化資料（t-SNE 對資料範圍敏感）
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. 初始化並執行 t-SNE，將高維度的資料降為 2 維
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

# 4. 視覺化結果
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap=plt.get_cmap('tab10'), s=50)
plt.colorbar(scatter)
plt.title('t-SNE on Digits Dataset')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_breast_cancer
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

# 1. 載入 breast_cancer 數據集
cancer = load_breast_cancer()
X = cancer.data  # 資料 (30 個特徵)
y = cancer.target  # 標籤 (0:惡性, 1:良性)

# 2. 標準化資料
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. 使用 PCA 將維度從 30 降到 10
pca = PCA(n_components=10)
X_pca = pca.fit_transform(X_scaled)

# 4. 使用 t-SNE 將 PCA 處理過的資料進一步降到 2 維
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_pca)

# 5. 視覺化結果
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap=plt.get_cmap('coolwarm'), s=50)
plt.colorbar(scatter, ticks=[0, 1], label='Class (0: Malignant, 1: Benign)')
plt.title('t-SNE after PCA on Breast Cancer Dataset')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()
```

## 使用技巧
1. 資料標準化：在應用 t-SNE 之前，先對資料進行標準化或歸一化，以避免數據尺度不同造成的影響。

2. 適當的參數選擇：
- perplexity（混亂度）：它控制了t-SNE中的鄰居數，通常範圍在5到50之間。較小的 perplexity 適合資料量較小或具明顯分群的資料，而較大的perplexity適合複雜或稀疏資料。
- 學習率（learning rate）：可以從較大的範圍（如10到1000）嘗試不同值。若結果分散得過於劇烈或重疊，可以調整學習率。

3. 避免過度解讀：t-SNE 是視覺化工具，主要用來發現資料中的趨勢和結構，但不應該用來嚴格量化資料之間的關係，尤其在解釋全局結構時需要小心。
4. 隨機性：t-SNE 有一定的隨機性，每次運行的結果可能會略有不同，因此可以使用固定的隨機種子來保持結果穩定。

降維的好處?

（1）加速計算速度，維度越少，計算量越少，並且能夠使用那些不適合於高維度的演算法

（2）去除一些冗餘的特徵(這邊的冗余可以是相關性太高的兩個 feature 或者是不是很重要的特徵)

    - 資料的多重共線性：特徵屬性之間存在著相互關聯關係。多重共線性會導致解的空間不穩定， 從而導致模型的泛化能力弱

（3）便於觀察和挖掘資訊（如將數據維度降到2維或者3維使之能可視化）

（4）特徵太多或者太複雜會使得模型過擬合

```python
import numpy as np
from scipy.stats import boxcox
from sklearn.model_selection import train_test_split

# 產生一組正值資料
data = np.random.exponential(scale=2, size=1000)

# 將資料分割為訓練和測試集
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

# 1. 對訓練數據應用 Box-Cox 並找到最佳的 λ 值
train_data_transformed, lambda_ = boxcox(train_data)

# 2. 將同樣的 λ 值應用到測試數據
test_data_transformed = boxcox(test_data, lmbda=lambda_)

# 2. 將同樣的 λ 值應用到測試數據
train_data_transformed_2 = boxcox(train_data, lmbda=lambda_)

# 結果
print(f"最佳的 λ 值: {lambda_}")
print(f"轉換後的訓練資料: {train_data_transformed[:5]}")
print(f"轉換後的訓練資料2: {train_data_transformed_2[:5]}")
print(f"轉換後的測試資料: {test_data_transformed[:5]}")
```
